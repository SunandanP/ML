```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("temperatures.csv")
df
```

```python
x = df.YEAR
x = x.values.reshape(117,1)
x
```

```python
y = pd.DataFrame({
    "Jan" : df.JAN,
    "Feb" : df.FEB,
    "Mar" : df.MAR,
    "Apr" : df.APR,
    "May" : df.MAR,
    "Jun" : df.JUN,
    "Jul" : df.JUL,
    "Aug" : df.AUG,
    "Sep" : df.SEP,
    "Oct" : df.OCT,
    "Nov" : df.NOV,
    "Dec" : df.DEC
})
```

```python
# reshaping the temperatures
for a in y:
    q = [z for z in y[a]]
    q = pd.Series(q)
    y[a] = q.values.reshape(117,1)
```

```python
y.Jan.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressors = []
for a in y:
    reg = LinearRegression()
    regressors.append(reg)
    reg.fit(x,y[a])
```
