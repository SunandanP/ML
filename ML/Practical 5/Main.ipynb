```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Mall_Customers.csv")
df
```

```python
df = df.iloc[:,3:]
df
```

```python
plt.title("Spending")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.scatter(df["Annual Income (k$)"], df["Spending Score (1-100)"], c="r")
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
```

```python
sse = []
for k in range(1,16):
    km = KMeans(n_clusters=k)
    km.fit_predict(df)
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.title("Elbow point")
plt.xlabel("No. of clusters")
plt.ylabel("SSE")
plt.xticks(range(1,16))
plt.grid()
plt.plot(range(1,16), sse , marker=".", color="r")
```

```python
# Silhouette Score way
from sklearn.metrics import silhouette_score
```

```python
ss = []
for k in range(2, 16):
    km = KMeans(n_clusters=k)
    labels = km.fit_predict(df)
    ss.append(silhouette_score(df, labels))
```

```python
ss
```

```python
plt.title("Silhouette Way")
plt.xticks(range(2,16))
plt.ylabel("Silhouette Score")
plt.xlabel("No of Clusters")
plt.bar(range(2,16), ss)
```

```python
km = KMeans(n_clusters=5)
```

```python
labels = km.fit_predict(df)
```

```python
labels
```

```python
centers = km.cluster_centers_
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title("Unclustered data")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score")
plt.scatter(df["Annual Income (k$)"], df["Spending Score (1-100)"], c="b")

plt.subplot(1,2,2)
plt.title("Clustered data")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score")
plt.scatter(df["Annual Income (k$)"], df["Spending Score (1-100)"], c=labels)

plt.scatter(centers[:,0], centers[:,1], s=100, color="k")
```

```python
ag = AgglomerativeClustering(n_clusters=5)
```

```python
alabels = ag.fit_predict(df)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title("Agglomerative")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score")
plt.scatter(df["Annual Income (k$)"], df["Spending Score (1-100)"], c=alabels)

plt.subplot(1,2,2)
plt.title("K-Means")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score")
plt.scatter(df["Annual Income (k$)"], df["Spending Score (1-100)"], c=labels)

plt.scatter(centers[:,0], centers[:,1], s=100, color="k")
```
