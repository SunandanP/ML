```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("Admission_Predict.csv")
df
```

```python
df.columns
```

```python
# Label encoding is not required here as the data in the frame is numeric data not string or anything

from sklearn.preprocessing import Binarizer
bi = Binarizer(threshold=0.75)

df["Chance of Admit "] = bi.fit_transform(df[["Chance of Admit "]])
df
```

```python
# set up of input and output data

x = df.drop("Chance of Admit ", axis=1)
y = df["Chance of Admit "]
```

```python
y
```

```python
y = y.astype(int)
y
```

```python
sns.countplot(x=y)
```

```python
y.value_counts()
```

```python
# Cross Validation

from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y = train_test_split(x,y, random_state=23, test_size=0.25)
train_x
```

```python
train_x
```

```python
train_x.shape
```

```python
test_x.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier()
```

```python
tree.fit(train_x, train_y)
```

```python
predicted = tree.predict(test_x)
```

```python
result = pd.DataFrame({
    'actual' : test_y,
    'predicted' : predicted
})
result
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(result.actual, result.predicted))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, accuracy_score
```

```python
ConfusionMatrixDisplay.from_predictions(result.actual, result.predicted)
```

```python
accuracy_score(result.actual, result.predicted)
```
